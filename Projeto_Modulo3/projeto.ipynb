```python
#fazendo várias requisições e concatenando os dataframes
df = pd.DataFrame()
for i in range(80):
    exemple = requests.get("http://localhost:3000/api/ep1")
    dados = exemple.json()
    df_aux = pd.DataFrame(dados) 
    df = pd.concat([df,df_aux], axis = 0)
```

```python
df
```

```python
#Transformando a data de milésimo de segundo para formato normal
df['date_converted'] = df['date'].apply(lambda x:datetime.datetime.fromtimestamp(x))
df
```

```python

df['date_converted'] = df['date_converted'].dt.strftime('%Y-%m-%d')
```

```python
#Substituindo os NaN por zero
df = df.fillna(0)
```

```python
#Pegando a data da ultima semana
maxima = max(df['date_converted'] )

data = datetime.datetime.strptime(maxima, "%Y-%m-%d") - datetime.timedelta(7)
data = datetime.datetime.strftime(data, "%Y-%m-%d")
data
```

```python
#Filtrando os produtos da última semana
df[df['date_converted'] >= data]
```

```python
#Cálculo da porcentagem sobre o produto novo
df_novos = df[df['date_converted'] >= data]
df_filtro = pd.DataFrame(df_novos[['prod_0', 'prod_1', 'prod_2', 'prod_3', 'prod_4', 'prod_5', 'prod_6', 'prod_7', 'prod_8', 'prod_9', 'prod_10', 'prod_11', 'prod_15']]*1.15)
```

```python

df_filtro_final = pd.concat([df_novos[['id', 'date', 'date_converted']], df_filtro], axis = 1)
```

```python
df[df['date_converted'] >= data] = df_filtro_final
```
